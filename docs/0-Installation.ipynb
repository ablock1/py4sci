## Anaconda (Python)

Anaconda is a cross platform package and environment manager. Meaning it will install Python and Conda package manager that will be using through this guide.

### Install

Follow the Installation procedure described bellow:

- **Windows**: https://docs.anaconda.com/anaconda/install/windows/  
- **Mac**: https://docs.anaconda.com/anaconda/install/mac-os/
- **Linux**:  https://docs.anaconda.com/anaconda/install/linux/

### Check your installation

Please verify your installation https://docs.anaconda.com/anaconda/install/verify-install/

In you command line (Anaconda prompt for Windows) type

```python
python
``` 
you should get into the python command line with a header similar to this
```python
Python 3.7.6 (default, Jan  8 2020, 19:59:22) 
[GCC 7.3.0] :: Anaconda, Inc. on linux
Type "help", "copyright", "credits" or "license" for more information.
```

where you verify the Python version "3.7.6" and the "Anaconda" manager. Quit with `exit()` or `Ctrl+D`.

## Jupyter Notebook

Notebooks are useful to to run codes in small cells rather than a whole script. [Jupyter](https://jupyter.org/) is the most common notebook for python and we are going to be use it during this guide. Each chapter (including this that you are reading) will be one individual notebook that you can download, run and modify locally. 

### Install

You already did it. Anaconda install Jupyter with its default packages in base enviroment.

### Launch

Using Terminal (Anaconda prompt for Windows), navigate to the folder where you want to open or create a notebook. Type:

```python
jupyter notebook
```

Don't close the Terminal while using Jupyter. Quit `Ctrl + C`.
### Cells

We are going to use two types of cells

1. **Code**: for Python codes
2. **Markdown**: For text comments* 

* for Markdown syntax https://www.markdownguide.org/basic-syntax/

### Commands

**Run a cell** : `Shift + Enter`

**New cell bellow** : `Esc + a`

**Delete a cell** : `Esc + d + d`

Complete list of commands in `Help > Keyboard Shortcuts`
