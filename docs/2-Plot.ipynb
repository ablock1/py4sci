## Multiple curves

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```


First download the online available data as described [here](http://localhost:8888/notebooks/docs/1-Basics.ipynb#Read-and-write-CSV-(Pandas)), then read the local CSV file.

```python
# Read CSV file
load_csv = pd.read_csv('CovidTotalCases.csv', delimiter=',')
# Show data
load_csv[-5:]
```

```python
# days strings will be packed on graph, lets create a x axis
xaxis = np.linspace(1,len(load_csv['date']),len(load_csv['date']))

# plot using matplotlib

# define font size
plt.rc('font', size=14) 

# define graph size
plt.figure(figsize=(10,7))

# plot many curves in same graph plot(X, Y, label)
plt.plot(xaxis, load_csv['Spain'], marker="X", label='Spain')
plt.plot(xaxis, load_csv['United Kingdom'], marker="X", label='UK')
plt.plot(xaxis, load_csv['Netherlands'], marker="X", label='Netherlands')
plt.plot(xaxis, load_csv['Germany'], marker="X", label='Germany')
plt.plot(xaxis, load_csv['South Korea'], marker="X", label='South Korea')

# Set axis labels and legends
plt.title('Covid-19 Total Cases - ourworldindata.org')
plt.xlabel('Time (days)')
plt.ylabel('# Detected Cases')
plt.legend(frameon=False)

# Add a text over graph text(pos x, pos y, text)
plt.text(5, 4000, "Last measured day: {!s}".format(load_csv['date'].iloc[-1]))

# Save as PNG image 
# plt.savefig('CovidTotalCases.png');

plt.plot();
```

## Multiple graphs

Let's plot multiple graphs on same figure.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```


First download the online available data as described [here](http://localhost:8888/notebooks/docs/1-Basics.ipynb#Read-and-write-CSV-(Pandas)), then read the local CSV file.

```python
# Read CSV file
load_csv = pd.read_csv('CovidTotalCases.csv', delimiter=',')

# days strings will be packed on graph, lets create a x axis
xaxis = np.linspace(1,len(load_csv['date']),len(load_csv['date']))
```

### Two graphs

```python
# define font size
plt.rc('font', size=14) 

# Set subplot layout
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14,5))

# plot graphs
ax1.plot(xaxis, load_csv['World'],"o-", color="royalblue", label='World')
ax2.plot(xaxis, load_csv['China'], "o-", color="darkorchid", label='China')

# set axis label
ax1.set_ylabel('# Total Cases')
ax1.set_xlabel('Time (days)')
ax2.set_xlabel('Time (days)')
ax1.legend(frameon=False)
ax2.legend(frameon=False)

# Save as PNG image 
# plt.savefig('CovidTotalCases_multgraphs.png');

plt.plot();
```

### Four graphs

```python
# define font size
plt.rc('font', size=14) 

# Set subplot layout
fig, ax = plt.subplots(2, 2, figsize=(12,8))

# plot graphs
ax[0,0].plot(xaxis, load_csv['World'],"o-", color="royalblue", label='World')
ax[0,1].plot(xaxis, load_csv['China'], "o-", color="darkorchid", label='China')
ax[1,0].plot(xaxis, load_csv['South Korea'],"o-", color="darkorange", label='South Korea')
ax[1,1].plot(xaxis, load_csv['Italy'], "o-", color="hotpink", label='Italy')

# set axis label
ax[0,0].set_ylabel('# Total Cases')
ax[1,0].set_ylabel('# Total Cases')
ax[1,0].set_xlabel('Time (days)')
ax[1,1].set_xlabel('Time (days)')

ax[0,0].legend(frameon=False)
ax[0,1].legend(frameon=False)
ax[1,0].legend(frameon=False)
ax[1,1].legend(frameon=False)

# Save as PNG image 
# plt.savefig('CovidTotalCases_multgraphs4.png');

plt.plot();
```

## Bars

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(font_scale=1.3)
sns.set_palette('tab20_r')
```


First download the online available data as described [here](http://localhost:8888/notebooks/docs/1-Basics.ipynb#Read-and-write-CSV-(Pandas)), then read the local CSV file.

```python
# Read CSV file
load_csv = pd.read_csv('CovidTotalCases.csv', delimiter=',')

# days strings will be packed on graph, lets create a x axis
xaxis = np.linspace(1,len(load_csv['date']),len(load_csv['date']))
```

Lets use **Seaborn** barplot to make a nice bar graph

```python
# Set a list of contries to plot
countries = ['United Kingdom', 'Netherlands', 'Germany', 'Spain', 'Italy', 'South Korea']

# This make a list of lats total number of cases for each country in countries
total_cases = [load_csv[x].iloc[-1] for x in countries] 

# plot
plt.figure(figsize=(8,5))

sns.barplot(y=countries, x=total_cases)

plt.title('Covid-19 Total Cases - ourworldindata.org')
plt.xlabel('# Detected Cases')

plt.plot();
```
